### 問1

文字列"datascience"の文字を逆に（末尾から先頭に向かって）並べた文字列を出力してください

```python
def reverse_string(input_string):
    return input_string[::-1]

original_string = "datascience"
reversed_string = reverse_string(original_string)

print(reversed_string)
```

### 問2

「ヘエルスシトンニキア」という文字列の1,3,5,7,9文字目を取り出して連結した文字列を出力してください


```python
text = "ヘエルスシトンニキア"
indices = [0, 2, 4, 6, 8]
result = ""

for i in indices:
    result += text[i]

print(result)
```

### 問3

「ヘルシンキ」＋「エストニア」の文字を先頭から交互に連結して文字列「ヘエルスシトンニキア」を出力してください


```python
str1 = "ヘルシンキ"
str2 = "エストニア"
result = ""

max_length = max(len(str1), len(str2))

for i in range(max_length):
    if i < len(str1):
        result += str1[i]
    
    if i < len(str2):
        result += str2[i]

print(result)
```

### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
sentence = "Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"

words = sentence.split()

word_lengths = [len(word) for word in words]

print(word_lengths)
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
import re

sentence = "It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."

cleaned_sentence = re.sub(r'[,.;:!?]', '', sentence)

words = cleaned_sentence.split()

print("分解された単語リスト:")
for i, word in enumerate(words, 1):
    print(f"{i}. {word}")

positions = [1, 2, 5, 6, 8, 9, 16]

result_dict = {}

for pos in positions:
    if pos <= len(words):
        word = words[pos-1]
        
        if len(word) >= 2:
            prefix = word[:2]
            result_dict[prefix] = pos

print("\n指定された位置の単語:")
for pos in positions:
    if pos <= len(words):
        print(f"{pos}. {words[pos-1]}")

print("\n結果の辞書:")
print(result_dict)
```
